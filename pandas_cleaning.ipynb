```python
import pandas as pd
```

```python
df = pd.read_excel('fAST patents v1 vs. v3 comparison.xlsx',sheet_name=0)
df = df.iloc[:,1:]
df.dropna(inplace=True)
df
```

```python
df['family_members'] = df['family_members'].str.split('|')
df = df.explode('family_members').reset_index().iloc[:,1:]
df['family_members'] = df['family_members'].str.strip()
df = df.drop_duplicates()
df = df.dropna(inplace=True)
```

```python
df.to_excel('final.xlsx',index=False)
```
